```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import spacy
import stanza
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
nlp = spacy.load('en_core_web_sm')
stanza.download('fr')
nlpf = stanza.Pipeline('fr')
nltk.download('popular')
```

```python

lang = input("Choose language : French, English")


def text_lemm_fr(phrase):
    doc = nlpf(phrase)
    text_lemm = [word.lemma for sentence in doc.sentences for word in sentence.words]
    txt_lemm_join = " ".join(text_lemm)
    return txt_lemm_join

def text_lemm(phrase) :
  phrase_tok = nlp(phrase)
  text_lemm = [token.lemma_ for token in phrase_tok]
  txt_lemm_join = " ".join(text_lemm)
  return txt_lemm_join

if (lang == "French"):
  your_input = input("Please enter a sentence")
  your_input = your_input.lower()
  choix_methode = input("Please choose the text_preprocessing method : Lemmatize, Stemming, TF-IDF, Wordcloud")


  tfidf = TfidfVectorizer()
  wordcloud = WordCloud()
  result_CV = ""
  df = pd.DataFrame()


  stem_en = SnowballStemmer("french")

  def fctn_stemming_fr (text):
    mots = word_tokenize(text, language="french")
    tokens_stemming  = [stem_en.stem(mot) for mot in mots]
    return ' '.join(tokens_stemming)

  if (choix_methode == "Lemmatize"):
      # Lemmatize
      print(text_lemm_fr(your_input))

  elif (choix_methode == "TF-IDF"):
      # TF-IDF
      your_input = [your_input]
      result_CV = tfidf.fit_transform(your_input)
      df = pd.DataFrame(result_CV.toarray(), columns=tfidf.get_feature_names_out())
      print(df)
  elif (choix_methode == "WordCloud"):
      # wordcloud à partir du texte
      wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10)
      wordcloud.generate_from_text(your_input)
      plt.figure()
      plt.imshow(wordcloud, interpolation="bilinear")
      plt.axis("off")
      plt.margins(x=0, y=0)
      plt.show()
  elif (choix_methode == "Stemming"):
    #Stemming
    print(fctn_stemming_fr(your_input))
  else :
    print("Error in the choice of the preprocessing method")







elif (lang == "English") :
  your_input = input("Please enter a sentence")
  your_input = your_input.lower()
  choix_methode = input("Please choose the text_preprocessing method : Lemmatize, Stemming, TF-IDF, Wordcloud")

  nlp = spacy.load('en_core_web_sm')
  tfidf = TfidfVectorizer()
  wordcloud = WordCloud()
  result_CV = ""
  df = pd.DataFrame()

  stem_en = SnowballStemmer("english")

  def fctn_stemming (text):
    mots = word_tokenize(text, language="english")
    tokens_stemming  = [stem_en.stem(mot) for mot in mots]
    return ' '.join(tokens_stemming)

  if (choix_methode == "Lemmatize"):
      # Lemmatize
      print(text_lemm(your_input))
  elif (choix_methode == "TF-IDF"):
      # TF-IDF
      your_input = [your_input]
      result_CV = tfidf.fit_transform(your_input)
      df = pd.DataFrame(result_CV.toarray(), columns=tfidf.get_feature_names_out())
      print(df)
  elif (choix_methode == "WordCloud"):
      # wordcloud à partir du texte
      wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10)
      wordcloud.generate_from_text(your_input)
      plt.figure()
      plt.imshow(wordcloud, interpolation="bilinear")
      plt.axis("off")
      plt.margins(x=0, y=0)
      plt.show()
  elif (choix_methode == "Stemming"):
    #Stemming
    print(fctn_stemming(your_input))
  else :
    print("Error in the choice of the preprocessing method")
```
